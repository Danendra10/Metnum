```python
def bisectionMethod(f, a, b, e):
    epsilon = np.abs(a-b)
    i = 0
    while True:
        c = (a+b)/2
        fa = f(a)
        fb = f(b)
        fc = f(c)
        epsilon = np.abs(a-b)
        print(i, a, b, c, fa, fb, fc, epsilon)
        if (fa*fc) < 0:
            b = c
        elif (fb*fc) < 0:
            a = c
        i+=1
        if epsilon < e: break
        if i > 1000:
            print ("tidak ada jawaban")
            break
    return c, epsilon
```

```python
def regulaFalsiMethod(f, a, b, e):
    epsilon = np.abs(a-b)
    i = 0
    while True:
        fa = np.round(f(a),6)
        fb = np.round(f(b),6)
        c = np.round(b - (fb*(b-a)/(fb-fa)), 6)
        fc = np.round(f(c),6)
        epsilon = np.abs(a-b)
        print(i, a, b, c, fa, fb, fc, epsilon)
        if (fa*fc) < 0:
            b = c
        elif (fb*fc) < 0:
            a = c
        i+=1
        if epsilon < e: break
        if i > 1000:
            print ("tidak ada jawaban")
            break
    return c, epsilon
```
